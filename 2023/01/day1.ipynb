# Load the input

```python
file = open('input.txt', 'r')
lines = file.readlines()
strings = []
for line in lines:
    strings.append(line.strip())
file.close()
```

# Find first and last digit

```python
import re
digits = []
for value in strings:
    first_digit = re.search("^[^\d]*(\d)", value).group(1)
    last_digit = re.search("(\d)[^\d]*$", value).group(1)
    print(first_digit, last_digit)
    digits.append((first_digit, last_digit))
```

# Find calibration value

```python
calibration_value = 0
for (first, last) in digits:
    value_to_add = int(first+last)
    print(value_to_add)
    calibration_value += value_to_add
```

# End of part 1

```python
print(calibration_value)
```

# Part 2 - Convert the spelled numbers

```python
spelled_numbers_plus_digit = "(one)|(two)|(three)|(four)|(five)|(six)|(seven)|(eight)|(nine)|(\d)"

for value in strings:
    first_digit = re.search(f"^(?!{spelled_numbers_plus_digit})*({spelled_numbers_plus_digit})", value).group(1)
    last_digit = re.search(f"({spelled_numbers_plus_digit})(?!{spelled_numbers_plus_digit})*$", value).group(1)
    print(first_digit, last_digit)
    digits.append((first_digit, last_digit))
```
